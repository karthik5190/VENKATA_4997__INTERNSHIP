# Data PreProcessing

```python
xtr=train_data.drop(['id','Seat comfort','Inflight entertainment','Ease of Online booking','Departure Delay in Minutes','satisfaction'],axis=1)
xte=test_data.drop(['id','Seat comfort','Inflight entertainment','Ease of Online booking','Departure Delay in Minutes'],axis=1)
ytr=train_data[['satisfaction']]
```

```python
xtr
```

```python
xte
```

```python
num_cols=list(xtr.select_dtypes(exclude='object').columns)
cat_cols=list(xtr.select_dtypes(include='object').columns)
```

```python
cat_cols
```

```python
num_cols
```

```python
xtr_num=xtr[num_cols]
xtr_cat=xtr[cat_cols]
xtr[cat_cols]=xtr[cat_cols].astype("category")
xtr_cat
```

```python
num_cols1=list(xte.select_dtypes(exclude='object').columns)
cat_cols1=list(xte.select_dtypes(include='object').columns)
```

```python
cat_cols1
```

```python
num_cols1
```

```python
xte_num=xte[num_cols1]
xte_cat=xte[cat_cols1]
xte[cat_cols1]=xte[cat_cols1].astype("category")
xte_cat
```

```python
from sklearn.impute import SimpleImputer 
xtr_imp=SimpleImputer(strategy='most_frequent')
xtr_imp.fit(xtr_cat)
xtr_cat=pd.DataFrame(xtr_imp.transform(xtr_cat),columns=cat_cols)
```

```python
xtr_imp=SimpleImputer(strategy='mean')
xtr_imp.fit(xtr_num)
xtr_num=pd.DataFrame(xtr_imp.transform(xtr_num),columns=num_cols)
```

```python
xtr_num.isnull().sum()
```

```python
xtr_cat.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer 
xte_imp=SimpleImputer(strategy='most_frequent')
xte_imp.fit(xte_cat)
xte_cat=pd.DataFrame(xte_imp.transform(xte_cat),columns=cat_cols1)
```

```python
xte_imp=SimpleImputer(strategy='mean')
xte_imp.fit(xte_num)
xte_num=pd.DataFrame(xte_imp.transform(xte_num),columns=num_cols1)
```

```python
xte_cat.isnull().sum()
```

```python
xte_num.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for col in cat_cols:
  xtr_cat[col]=encoder.fit_transform(xtr_cat[col])
```

```python
xtr_cat
```

```python
train_data=pd.concat([xtr_cat,xtr_num],axis=1)
```

```python
train_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for col in cat_cols1:
  xte_cat[col]=encoder.fit_transform(xte_cat[col])
```

```python
xte_cat
```

```python
test_data=pd.concat([xte_cat,xte_num],axis=1)
```

```python
test_data.head()
```

```python
columns=train_data.columns
columns
```

```python
columns1=test_data.columns
columns1
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
train_data=pd.DataFrame(scale.fit_transform(train_data),columns=columns)
test_data=pd.DataFrame(scale.fit_transform(test_data),columns=columns1)
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data,ytr,test_size=0.2,random_state=4624)
print(x_train.shape)
print(y_train.shape)
y_test.shape
```

```python
from sklearn import linear_model
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import GridSearchCV
logreg = linear_model.LogisticRegression(random_state = 42)
```

```python
param_grid = {'penalty': ['l1', 'l2'],
              'C': [0.001, 0.01, 0.1, 1, 10, 100],
              'max_iter':[100,500,1000,5000],
              'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
```

```python
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)
```

```python
print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Accuracy Score: ", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
print("Validation Accuracy Score: ", best_model.score(x_test, y_test))
```

```python
print("test accuracy: {} ".format(logreg.fit(x_train, y_train).score(x_test, y_test)))
print("train accuracy: {} ".format(logreg.fit(x_train, y_train).score(x_train, y_train)))
```
